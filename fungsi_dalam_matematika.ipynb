Sebelum lebih jauh membahasa mengenai fungsi dalam matematika, ada satu definisi yang harus dipahami terlabih dahulu yaitu **relasi**. Misal $A$ dab $B$ adalah himpunan maka $A$ dan $B$ dikatakan memiliki **relasi** jika terdapat anggota himpunan yang saling berpasangan. Contoh $A=\{\text{kucing},\text{kambing},\text{kuda}\}$ dan $B=\{\text{rumput},\text{pizza},\text{whiskas}^{\small{TM}}\}$ misal relasi antara $A$ dan $B$ adalah "menyukai" maka diagram di bawah ini mengilustrasikan hal tersebut

relasi.png

Fungsi $f$ dari $A$ ke $B$ didefinisikan sebagai relasi dari himpunan $A$ ke himpunan $B$ dinotasikan $f:A\to B$, jika setiap anggota himpunan $A$ berpasangan tepat satu dengan anggota himpunan $B$. Disini perlu diperhatikan bahwa setiap anggota $A$ tidak boleh berhubungan dengan lebih dari satu anggota $B$ tetapi tidak berlaku sebaliknya, anggota $B$ boleh berpasangan dengan lebih dari satu anggota $A$. Misal $x\in A$ ($x$ anggota himpunan $A$) maka $x$ dipetakan oleh $f$ dinotasikan sebagai $f(x)$, untuk kasus diatas $f(\text{kuda})=\text{rumput}$. Himpunan $A$ disebut domain dan himpunan $B$ disebut kodomain, dan subhimpunan dari $B$ yang merupakan hasil pemetaan dari A disebut range, atau dinotasikan sebagai $\text{range}(f)=\{f(x)|x\in A\}$. Fungsi dalam matematika biasa ditulis menggunakan persamaan misal $f:A\to B$, $f(x)=2x$, untuk menggambar fungsi dalam digram kartesius, kita akan mencacah setiap pasangan $(x,f(x))$ untuk setiap $x\in A$ kemudian merepresentasikannya sebagai sebuah titik pada diagram kartesius. Dalam bahasa pemrograman python terdapat beberapa pustaka pihak ketiga yang dapat digunakan untuk menggambar fungsi matematika, salah satunya adalah matplotlip

```python
import matplotlib.pyplot as plt
import numpy as np

A = np.linspace(0.1,10,100)
A
```

`np.linspace(0.1,10,100)` akan membuat array yang berisi angka dimulai dari $0.1$ dan akan ditambah $\frac{10}{100}$ sebanyak 100 kali sampai mencapai nilai $10$


```python
def f(x):
  return 2*x
```

```python
range_f = f(A)
plt.plot(A,range_f)
plt.show()
```

contoh lain misal $g(x) = x^3-2$

```python
def g(x):
  return x**3-2
```

```python
Xs = np.linspace(-10,10,100)
range_g = g(Xs)
plt.plot(Xs,range_g)
plt.show()
```

input pertama dan kedua untuk `plt.plot()` berturut-turut adalah domain dan range untuk $f$. Untuk kasus fungsi yang memiliki 2 peubah seperti $f(x,y)=x^2+y^2$, misal domain untuk $f$ adalah $G$ dan $H$, analog dengan kasus 1 peubah kita akan mencacah setiap titik $(x,y,f(x,y))$ untuk setiap $x\in G$ dan $y\in H$ tetapi kali ini dalam ruang 3 dimensi. Matplotlib menyediakan berbagai jenis style dari gambar dalam contoh ini kita akan menggunakan kontur 3 dimensi 

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt


def f(x, y):
    return (x ** 2 + y ** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
print(x)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)



fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
